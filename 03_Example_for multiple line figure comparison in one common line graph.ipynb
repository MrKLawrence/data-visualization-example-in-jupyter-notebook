```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# %load get_df
def get_df(yr):
    return pd.read_csv("Environmental_Data_Deep_Moor_{}.csv".format(yr))
#Note that you can load a Python file using "%load" magic command without giving the extension but just the relative path
```

```python
# %load line_helpers.py
#This calculates an average of a given column in the DataFrame for a given month
def monthly_avg_calc(mo,col):
    return df[df['date'].str.contains('201[2345]_[0]?'+ str(mo))][col].mean()
```

```python
#This explicitly define a DataFrame to be used so I read it in just once for this exercise
def yearly_avg(category):
    return list(map(lambda m: monthly_avg_calc(m, category),range(1,13)))
```

```python
df = get_df('2014')
```

```python
#Here plots the average wind speed over the course of 2014
plt.plot(yearly_avg('Wind_Speed'))
plt.show()
```

```python
#We plot multiple figures in the same line graph here to see the correlations by calling the plot function multiple times
plt.plot(yearly_avg('Wind_Speed'))
plt.plot(yearly_avg('Wind_Gust'))
plt.plot(yearly_avg('Dew_Point'))
plt.plot(yearly_avg('Barometric_Press'))
plt.show()
#The result below is interesting as those 2 lines at the bottom (in green colour) look strikingly similar and the one at the top looks like it has a strong inverse relationship to the ones at the bottom.
```

```python
#add a label for each line
plt.plot(yearly_avg('Wind_Speed'),label='Wind Speed')
plt.plot(yearly_avg('Wind_Gust'),'d-',label='Wind Gust')
plt.plot(yearly_avg('Dew_Point'),'d-',label='Dew Point')
plt.plot(yearly_avg('Barometric_Press'),label='Pressure')
plt.legend()
plt.show()
#This shows an inverse relationship between wind gust and dew point.
```

```python
plt.plot(yearly_avg('Wind_Speed'),label='Wind Speed',color="#2BD31444") #here we use the RGB(A) Hex Code in this instance to give a stronger alpha value
plt.plot(yearly_avg('Wind_Gust'),'d-',label='Wind Gust',color="#2BD314FF")
plt.plot(yearly_avg('Dew_Point'),'d-',label='Dew Point',color="#142BD3FF")
plt.plot(yearly_avg('Barometric_Press'),label='Pressure',color="#142BD366")
plt.legend()
plt.show()
```
